````python
#### SIMPLE EXERCISE:
## Write a function that takes a single string as input and calculates the GC content of it
## GC content is the percentage (or fraction) of a sequence that it is G or C.

def gc_calc(seq):
  count_c = seq.count('C')
  count_g = seq.count('G')
  count_t = seq.count('T')
  count_a = seq.count('A')
  return (count_g + count_c) / (count_c + count_g + count_a + count_t)

### run your function using sequence1 as input.

gc_cont = gc_calc(sequence1)
print(gc_cont)
````

Loops are essential constructs in programming that allow you to repeat a block of code multiple times. Python provides two main types of loops: for loops and while loops. These loops enable you to automate repetitive tasks, iterate through sequences, and process data efficiently.

fruits = ["apple", "banana", "cherry"]

count = 0

while count < 5:

    count += 1
